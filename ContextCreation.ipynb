```python
import pandas as pd
import glob
from keras.preprocessing.text import Tokenizer, text_to_word_sequence

path = r'drive/MyDrive/samples'
all_files = glob.glob(path + "/val*.csv")
all_files += glob.glob(path + "/t*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)

```

```python
frame = frame.sample(frac=1)

samples = " <EOF> <BOF> ".join(list(frame["sample"])).split(" ")
```

```python
import keras
import json

with open("drive/MyDrive/samples/tokenizer.json") as f:
  json_string = json.load(f)

tokenizer = keras.preprocessing.text.tokenizer_from_json(
    json_string
)

frame.head()
```

```python
import numpy as np

i=1

word_encodings = {}

for word in set(samples):
    word = word.strip()
    print(i)
    i += 1
    #print(np.where(tokenizer.texts_to_matrix([word]) == 1)[1])
    
    if len(np.where(tokenizer.texts_to_matrix([word]) == 1)[1]) == 1:
      word_encodings[word] = np.where(tokenizer.texts_to_matrix([word]) == 1)[1][0]
    else:
      word_encodings[word] = 1
```

```python
print(len(set(word_encodings.values())))
```

```python
from scipy import sparse
import time

count = 1

# for idx, row in frame.iterrows():
#   print(len(row["sample"]))
#   row["token"] = tokenizer.texts_to_matrix([row["sample"]], mode="count")
#   if count == 1:
#     print(row["sample"])
#     print("token shape is ", row["token"].shape)
  
frame["token"] = frame["submission_id"]
tokens = []

for index,row in frame.iterrows():
    #print(row["sample"].split(" "))
    now = time.time()
    token = sparse.csr_matrix(tokenizer.texts_to_matrix([row["sample"].split(" ")], mode="count"))
    if count == 1:
      print(token)
    #print("tokenized...")
    #now = time.time()
    tokens.append(token)
    #frame.loc[index, 'token'] = np.array2string(token)
    #then = time.time()
    count += 1
    print(count)
```

```python
frame["token"] = tokens
```
